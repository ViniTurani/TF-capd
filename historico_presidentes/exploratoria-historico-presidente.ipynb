```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

try:
    data = pd.read_csv("historico_presidentes/historico_presidentes.csv", delimiter=';', encoding='latin1')
except UnicodeDecodeError:
    try:
        data = pd.read_csv("historico_presidentes/historico_presidentes.csv", delimiter=';', encoding='iso-8859-1')
    except UnicodeDecodeError:
        data = pd.read_csv("historico_presidentes/historico_presidentes.csv", delimiter=';', encoding='cp1252')

print(data.head())

print(data.info())

print(data.isnull().sum())


```

```python
for column in data.columns:
    try:
        if not column.startswith("QT") and not column.endswith("ACUMULADO"):
            print(data[column].agg(['mean', 'median', 'std', 'min', 'max']))
    except:
        pass
  
```

```python
none_values = data.isna().sum()
for column in none_values:
    if column > 0:
        print(column)
```

```python
columns_to_drop = ['DT_TOTALIZACAO', 'PE_SECOES_TOT_ACUMULADO', 'CIRO_GOMES_PE_VOTOS_TOT_ACUMULADO', 
                'LULA_PE_VOTOS_TOT_ACUMULADO', 'PADRE_KELMON_PE_VOTOS_TOT_ACUMULADO', 
                'SIMONE_TEBET_PE_VOTOS_TOT_ACUMULADO', 'VERA_PE_VOTOS_TOT_ACUMULADO', 
                'SOFIA_MANZANO_PE_VOTOS_TOT_ACUMULADO', 'JAIR_BOLSONARO_PE_VOTOS_TOT_ACUMULADO', 
                'CONSTITUINTE_EYMAEL_PE_VOTOS_TOT_ACUMULADO', 'FELIPE_DAVILA_PE_VOTOS_TOT_ACUMULADO', 
                'SORAYA_THRONICKE_PE_VOTOS_TOT_ACUMULADO', 'LEO_PERICLES_PE_VOTOS_TOT_ACUMULADO', 
                'BRANCO_PE_VOTOS_TOT_ACUMULADO', 'NULO_PE_VOTOS_TOT_ACUMULADO', 'QT_SECOES_TOTAL', 
                'QT_SECOES_TOTAL', 'QT_APTOS_TOTAL', 'QT_SECOES_TOT', 'QT_SECOES_TOT_ACUMULADO', 'PE_SECOES_TOT_ACUMULADO',
                 'VOTOS_TOT_ACUMULADO', 'PE_VOTOS_TOT_ACUMULADO' , 'QT_VOTOS_TOT_ACUMULADO', 'DT_TOTALIZACAO', 'QT_VOTOS_TOTAL_ACUMULADO', 'QT_VOTOS_CONCORRENTES',
                 'QT_APTOS_TOT_ACUMULADO', 'ACUMULADO']

columns_to_drop_filtered = []
for col in columns_to_drop:
    if col in data.columns:
        columns_to_drop_filtered.append(col)
for col in data.columns:
    if any(col.startswith(prefix) for prefix in columns_to_drop):
        columns_to_drop_filtered.append(col)
    if any(col.endswith(suffix) for suffix in columns_to_drop):
        columns_to_drop_filtered.append(col)

data_cleaned = data.drop(columns=columns_to_drop_filtered)
data_cleaned = data_cleaned.fillna(0)
data_cleaned.to_csv("historico_presidentes_cleaned.csv", index=False)
```
